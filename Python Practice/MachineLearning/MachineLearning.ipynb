```python
# Basic Neural Networks
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
observations = 1000
xs=np.random.uniform(low=-10,high=10,size=(observations,1))
zs=np.random.uniform(-10,10,(observations,1))

inputs = np.column_stack((xs,zs))
xs.shape
```

```python
inputs.shape
```

```python
# targets = f(x,z) = 2*x - 3*z + 5 + noise
# noise is there to randomize data. B/c data is never perfect

noise = np.random.uniform(-1,1,(observations,1))
targets = 2*xs - 3*zs + 5 + noise
targets.shape
```

```python
targets = targets.reshape(observations,)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs[:,0],zs[:,0],targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')
ax.view_init(azim=100)
plt.show()
targets = targets.reshape(observations,1)
```

```python
init_range = 0.1
weights = np.random.uniform(-init_range, init_range, size=(2,1))
weights
```

```python
biases = np.random.uniform(-init_range,init_range,size=1)
biases
```

```python
learning_rate = 0.02 # eta
```

```python
# Now to train the model
for i in range(100): # bias will imporve when you increase the number of iterations
    outputs =  np.dot(inputs,weights) + biases
    deltas = outputs - targets
    loss=np.sum(deltas ** 2)/ 2 / observations
    print(loss) # if the numbers become lower and lower thats good
    # we're minimizing the loss function

    deltas_scaled=deltas / observations

    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)
```

```python
print(weights, biases)
```

```python
plt.plot(outputs, targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show() # we want the line to be a 45 degree. That is very good
```
